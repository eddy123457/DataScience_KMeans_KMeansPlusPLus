```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import KMeans
import numpy as np
df = pd.read_csv ('Mall_Customers.csv')
df2=df.rename({'CustomerID': "Customer", 'Annual Income (k$)': "Income" ,'Spending Score (1-100)' : 'Spending' }, axis = 'columns').replace(['Male'], 1).replace(['Female'], 0)

print(df2)
x=np.asarray(df.Age)
y=np.asarray(df2.Income)
df=np.array(list(zip(x, y))).reshape(len(y), 2)
kmeans = KMeans(n_clusters=4).fit(df)
centers = np.array(kmeans.cluster_centers_)
print(kmeans.cluster_centers_)
centers = np.array(kmeans.cluster_centers_)
plt.title('k means Algo')
plt.scatter(x, y, c=kmeans.labels_.astype(float))    
plt.scatter(centers[:,0], centers[:,1], marker="*", color='r', label= "Centroids")
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from pandas import DataFrame

Data = {
    'x': 
[35,34,32,37,33,33,31,27,35,34,62,54,57,47,50,57,59,52,61,47,50,48,39,40,45,47,39,44,50,48],
        'y': 
[79,54,52,77,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,23,22,13,14,22,7,29,25
,9,8]
       }
  
df = DataFrame(Data)
print(df)

kmeans = KMeans(n_clusters=3, init = 'k-means++').fit(df) 

centroidsK = kmeans.cluster_centers_ 
labelsK = kmeans.labels_ 

print(f'Centroids:\n{centroidsK}')
print()
print(f'Labels:\n{labelsK}')

df2 = {'x' : [34, 48, 62, 35],
       'y' : [52, 6, 49, 77]}

df2 = DataFrame(df2)

kmeans2 = kmeans.predict(df2)
print(kmeans2)

plt.scatter(df['x'], df['y'], c=kmeans.labels_) 
plt.plot(df2['x'], df2['y'], 'b+', markersize=10,label='Predicted')
plt.scatter(centroidsK[:, 0], centroidsK[:, 1], c='red', label = 'centroids', marker="*")
plt.legend()
plt.show()
```
